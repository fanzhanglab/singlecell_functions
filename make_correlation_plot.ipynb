## Make a correlation plots of two vectors

```r
library(ggpubr)
```

```r
getwd()
```

```r
exp <- readRDS("Psoriasis_gene_expression.RDS")
res <- readRDS("stCCA_Psoriasis_stccaobj_Spe1.RDS")
class(res)

# Low-dimensional embeddings, here these are stCC
stCC <- res$joint_emb

gene_list <- c("CD52", "KRT2")
combine <- cbind(stCC[, c(1:5)], exp[, which(colnames(exp) %in% gene_list)])

dim(combine)
combine[1:4, ]
```

```r
options(repr.plot.height = 5, repr.plot.width = 5)
ggscatter(combine, x = "stCC3", y = "KRT2",
          color = "white", shape = 21, size = 1, 
          add = "loess", cor.coef.size = 7,
          add.params = list(color = "blue", fill = "lightgray"),
          conf.int = TRUE,
          cor.coef = TRUE) +
          font("ylab", size = 20, face = "italic") +
          theme_classic(base_size = 25)
# ggsave("KRT2.pdf", width = 4, height = 4)

ggscatter(combine, x = "stCC3", y = "KRT2",
          color = "white", shape = 21, size = 1, 
          add = "reg.line", cor.coef.size = 7,
          add.params = list(color = "blue", fill = "lightgray"),
          conf.int = TRUE,
          cor.coef = TRUE) +
          font("ylab", size = 20, face = "italic") +
          theme_classic(base_size = 25)
# ggsave("KRT2.pdf", width = 4, height = 4)
```
